```python
# Support Vector Machine
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
url = "https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv"
iris = pd.read_csv(url)
iris.head()
```

```python
iris4 = iris.drop(['species'],axis='columns')
iris4.head()
```

```python
iris1 = iris4[:50]
iris2 = iris4[50:100]
iris1.head()
```

```python
iris2.head()
```

```python
plt.scatter(iris1.sepal_length,iris1.sepal_width,color='r')
plt.scatter(iris2.sepal_length,iris2.sepal_width,color='g')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(iris.drop('species',axis=1),iris['species'],train_size=0.9)
X_test
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
```
