```python
import pandas as pd
url = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
iris = pd.read_csv(url)
iris
iris1 = iris[:50] #setosa
iris2 = iris[50:100] #verosika
iris3 = iris[100:150]
plt.scatter(iris1.sepal_length,iris1.sepal_width)
plt.scatter(iris2.sepal_length,iris2.sepal_width)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(iris.drop('species',axis=1),iris['species'],train_size=0.9)
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```
