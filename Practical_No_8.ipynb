```python
 import pandas as pd
 from sklearn.ensemble import AdaBoostClassifier
 from sklearn.preprocessing import LabelEncoder
 from sklearn.tree import DecisionTreeClassifier
 from sklearn import metrics
 url="https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/mushrooms.csv"
 dataset=pd.read_csv(url)
 dataset.head()
```

```python
 dataset.columns=['type','cap_shape','cap_surface','cap_color','bruises','odor','gill_attachment','gill_spacing','gill_size','gill_color','stalk_shape','stalk_root','stalk_surface_above_ring','stalk_surface_below_ring','stalk_color_above_ring','stalk_color_below_ring','veil_type','veil_color','ring_number','ring_type','spore_print_color','population','habitat']
```

```python
 for label in dataset.columns: dataset[label]=LabelEncoder().fit(dataset[label]).transform(dataset[label])
```

```python
dataset.info()
```

```python
print(dataset.info())
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'dataset' is your DataFrame
X = dataset.drop(['type'], axis=1)
Y = dataset['type']

# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=1)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# Assuming you've already imported train_test_split and split your data into X_train, X_test, Y_train, Y_test

# Initialize a base model (a weak learner)
base_model = DecisionTreeClassifier()

# Initialize AdaBoost classifier
AdaBoost = AdaBoostClassifier(base_estimator=base_model, n_estimators=400, learning_rate=1)

# Train the AdaBoost model
boost_model = AdaBoost.fit(X_train, Y_train)

# Predictions
y_pred = boost_model.predict(X_test)

# Calculate accuracy
accuracy = metrics.accuracy_score(Y_test, y_pred)

# Print accuracy
print("The Accuracy IS : ", accuracy * 100, '%')
```
